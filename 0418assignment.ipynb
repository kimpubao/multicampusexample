```python
#2. 
'''
strip 함수를 구현하세요.
'''
test_string = "  Hello, world!  "
def custom_strip(input_string):
    start = 0
    end = len(input_string) - 1

    # 문자열의 시작부터 공백이 아닌 문자가 나올 때까지 인덱스를 이동
    while start <= end and input_string[start].isspace():
        start += 1

    # 문자열의 끝부터 공백이 아닌 문자가 나올 때까지 인덱스를 이동
    while end >= start and input_string[end].isspace():
        end -= 1

     # 시작과 끝을 기준으로 공백을 제거한 문자열을 반환
    return input_string[start:end+1]

print(custom_strip(test_string))
```
